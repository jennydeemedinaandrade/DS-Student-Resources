```r
# Create a histogram with suitable bin widths for the number of absences each student has
library("ggplot2")
```

```r
# Read in Data
student.mat = read.csv("../../Data/student-mat.csv")
```

```r
# View column names
names(student.mat)
```

```r
# View Data
head(student.mat)
```

```r
# Check dimensions of dataframe
dim(student.mat)
```

```r
# Histogram in ggplot
ggplot(student.mat, aes(x=absences)) + geom_histogram(bins=50) + ggtitle("Number of Student Absences") + xlab("Absences") + ylab("Frequency Count")
```

```r
# Create a boxplot of the number of absences each student has
ggplot(student.mat, aes(y=absences)) + geom_boxplot(outlier.color="red") + ggtitle("Number of Student Absences") + xlab("Absences") + ylab("Frequency Count")
```

```r
# A basic box with the schools colored
ggplot(student.mat, aes(x=school, y=absences, fill=school)) + geom_boxplot(outlier.color="red")
```

# Are there outliers? High or low?

# Yes there are outliers- they are all high # Create a normal probability plot

```r
# qq plot with ggplot
ggplot(student.mat, aes(sample=absences)) + geom_qq() +
ggtitle("Number of Student Absences") + xlab("Absences") + ylab("Frequency Count")
```
